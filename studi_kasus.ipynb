```python
import pandas as pd

df = pd.read_csv("data_kuesioner.csv")
df.head()

import matplotlib.pyplot as plt
import numpy as np
```

## 1. Bar Chart

Digunakan untuk memperlihatkan distribusi jawaban kusioner dari keseluruhan data.

```python
# Data kategori jawaban kuesioner
kategori = ['STS', 'TS', 'CS', 'S', 'SS']
jumlah = [12, 18, 25, 30, 15]  # contoh total jawaban

# Membuat Bar Chart
plt.bar(kategori, jumlah)

# Memberi judul dan label
plt.title('Distribusi Jawaban Kuesioner')
plt.xlabel('Kategori Jawaban')
plt.ylabel('Jumlah Jawaban')

# Menampilkan grafik
plt.show()
```

## 2. Pie Chart

Digunakan untuk memperlihatkan proporsi jawaban kusioner dari keseluruhan data.

```python
# Data kategori jawaban kuesioner
kategori = ['STS', 'TS', 'CS', 'S', 'SS']
jumlah = [12, 18, 25, 30, 15]  # contoh total jawaban

# Membuat Pie Chart
plt.pie(jumlah, labels=kategori, autopct='%1.1f%%', startangle=90)

# Memberi judul
plt.title('Proporsi Jawaban Kuesioner')

# Membuat pie berbentuk lingkaran
plt.axis('equal')

# Menampilkan grafik
plt.show()
```

## 3. Stacked bar

Digunakan untuk memperlihatkan distribusi jawaban per pertanyaan.

```python
# Data pertanyaan
pertanyaan = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']

# Distribusi jawaban per pertanyaan
STS = [2, 1, 3, 2, 1]
TS  = [3, 2, 2, 3, 2]
CS  = [5, 4, 3, 4, 3]
S   = [6, 7, 6, 5, 7]
SS  = [4, 6, 6, 6, 7]

# Membuat stacked bar
plt.bar(pertanyaan, STS, label='STS')
plt.bar(pertanyaan, TS, bottom=STS, label='TS')
plt.bar(pertanyaan, CS, bottom=np.array(STS) + np.array(TS), label='CS')
plt.bar(pertanyaan, S, bottom=np.array(STS) + np.array(TS) + np.array(CS), label='S')
plt.bar(pertanyaan, SS, bottom=np.array(STS) + np.array(TS) + np.array(CS) + np.array(S), label='SS')

# Memberi judul dan label
plt.title('Distribusi Jawaban Kuesioner per Pertanyaan')
plt.xlabel('Pertanyaan')
plt.ylabel('Jumlah Jawaban')

# Menampilkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```

## 4. Bar Chart

Digunakan untuk memperlihatkan rata-rata skor per pertanyaan.

```python
# Data pertanyaan
pertanyaan = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']

# Rata-rata skor per pertanyaan (contoh hasil perhitungan)
rata_rata = [3.8, 4.1, 3.6, 3.9, 4.2]

# Membuat Bar Chart
plt.bar(pertanyaan, rata_rata)

# Memberi judul dan label
plt.title('Rata-Rata Skor per Pertanyaan')
plt.xlabel('Pertanyaan')
plt.ylabel('Rata-Rata Skor')

# Menampilkan grafik
plt.ylim(0, 5)  # karena skala Likert 15
plt.show()
```

## 5. Bar Chart

Digunakan untuk memperlihatkan distribusi kategori jawaban positif, netral dan negatif.

```python
# Kategori sentimen jawaban
kategori = ['Negatif', 'Netral', 'Positif']

# Jumlah jawaban per kategori (contoh hasil pengelompokan)
jumlah = [20, 25, 45]

# Membuat Bar Chart
plt.bar(kategori, jumlah)

# Memberi judul dan label
plt.title('Distribusi Kategori Jawaban Kuesioner')
plt.xlabel('Kategori Jawaban')
plt.ylabel('Jumlah Jawaban')

# Menampilkan grafik
plt.show()
```

## 6. Line Plot

Digunakan untuk istribusi Jawaban per Pertanyaan.

```python
# Daftar pertanyaan
pertanyaan = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']

# Distribusi jawaban per kategori (contoh data)
STS = [2, 1, 3, 2, 1]
TS  = [3, 2, 2, 3, 2]
CS  = [5, 4, 3, 4, 3]
S   = [6, 7, 6, 5, 7]
SS  = [4, 6, 6, 6, 7]

# Membuat line plot
plt.plot(pertanyaan, STS, marker='o', label='STS')
plt.plot(pertanyaan, TS, marker='o', label='TS')
plt.plot(pertanyaan, CS, marker='o', label='CS')
plt.plot(pertanyaan, S, marker='o', label='S')
plt.plot(pertanyaan, SS, marker='o', label='SS')

# Judul dan label
plt.title('Distribusi Jawaban per Pertanyaan')
plt.xlabel('Pertanyaan')
plt.ylabel('Jumlah Jawaban')

# Menampilkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```
